1. Factorial:

    5! = 5 * 4 * 3 * 2 * 1

```python
# Factorial 

def factorial(n):
    fact = 1 
    for i in range(2,n+1):
        fact *= i 
    return fact 

factorial(5)
```

2. Fibonacci Numbers



```python
def printFibonacci(n):
    prev, cur, fibb = 1,1,0

    for i in range(2,n):
        fibb = prev + cur
        prev = cur 
        cur = fibb
    
    return fibb


printFibonacci(9)
```

## Search:

### 4.1  Linear Search

```python
#Search


def linear_search(arr,target):
    for i in range(len(arr)):
        if arr[i]==target:
            return i
    return -1

linear_search([23,45,23,22,21,12],23)
```

## 4.2 Binary Search



```python
#Binary Search

# Only for sorted arrays


def binary_search(arr,target):

    low,high = 0, len(arr)-1
    

    while high >= low :

        mid = low+high//2

        if arr[mid]==target:
            return mid
        elif target>arr[mid]:
            low = mid+1
        else:
            high = mid-1
    
    return -1



arr = [23,45,23,22,21,12]
arr.sort()
binary_search(arr,12)

```

## 5. Sorting Algoithms

### 5.1 Bubble Sort



 - At each pass largest element will sorted at the end
 - At each pass if No elements are Swapped, then the Array is Already Sorted

```python
def bubble_sort(arr):

    for i in range(len(arr)-1):
        swapped = False
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1] ,arr[j]   # Swap elements 
                swapped = True
                
        if swapped == False:
            break     # During First pass if there is no Swapping of Elements, then the given Array is already sorted
    return arr


arr = [5,3,55,2,1,12]
print(bubble_sort(arr))     
```

### 3. convert number to specific mapping 
 
  - Remember these asciii values 
    - a:65
    - b: 97
  - print(chr(65))  ## outputs: a
  - print(chr(97)) ## outputs: b


```python
#  map all the numbers --> 12345 --- abcde


def convertBarcode(n):
    a = str(n)
    ans = ""
    for i in range(len(a)):
        lastDigit = n%10
        ans += chr(lastDigit + 64)
        n = n//10
    return ans[::-1].lower()


print(convertBarcode(123444))
    
```

```python
number = 1234
string = ""

while number > 0:
    remainder = number % 26
    print(remainder)
    character = chr(remainder + 97)  # Add 97 to get the ASCII value of lowercase letters
    string = character + string
    number = number // 26

print(string)  # Output: "abcd"
```
