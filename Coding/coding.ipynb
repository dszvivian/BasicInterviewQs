1. Factorial:

    5! = 5 * 4 * 3 * 2 * 1

```python
# Factorial 

def factorial(n):
    fact = 1 
    for i in range(2,n+1):
        fact *= i 
    return fact 

factorial(5)
```

2. Fibonacci Numbers



```python
def printFibonacci(n):
    prev, cur, fibb = 1,1,0

    for i in range(2,n):
        fibb = prev + cur
        prev = cur 
        cur = fibb
    
    return fibb


printFibonacci(9)
```
