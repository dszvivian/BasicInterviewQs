1. Factorial:

    5! = 5 * 4 * 3 * 2 * 1

```python
# Factorial 

def factorial(n):
    fact = 1 
    for i in range(2,n+1):
        fact *= i 
    return fact 

factorial(5)
```

2. Fibonacci Numbers



```python
def printFibonacci(n):
    prev, cur, fibb = 1,1,0

    for i in range(2,n):
        fibb = prev + cur
        prev = cur 
        cur = fibb
    
    return fibb


printFibonacci(9)
```

## Search:

### 4.1  Linear Search

```python
#Search


def linear_search(arr,target):
    for i in range(len(arr)):
        if arr[i]==target:
            return i
    return -1

linear_search([23,45,23,22,21,12],23)
```

## 4.2 Binary Search



```python
#Binary Search

# Only for sorted arrays


def binary_search(arr,target):

    low,high = 0, len(arr)-1
    

    while high >= low :

        mid = low+high//2

        if arr[mid]==target:
            return mid
        elif target>arr[mid]:
            low = mid+1
        else:
            high = mid-1
    
    return -1



arr = [23,45,23,22,21,12]
arr.sort()
binary_search(arr,12)

```
